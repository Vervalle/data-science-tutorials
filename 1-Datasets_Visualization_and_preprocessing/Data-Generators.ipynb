# Data Generators
Here we briefly present two scikit-learn generators, namely [make_blobs](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs) and [make_moons](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html#sklearn.datasets.make_moons) to generate artifical data for clustering or classification. 
For further data generators, see the [documentation](http://scikit-learn.org/stable/datasets/index.html#sample-generators).

## Make_blobs
This method generates isotropic Gaussian blobs for clustering. Let's generate some datapoints, using the default values. This will generate 100 points equally divided along three clusters. The cluster centers are chosen randomly in the range [-10,10]. Standard deviation is set to 1.0. We explicitly set the random_state to 0, for reproducible data generation. If no random_state is provided, np.random is used, producing different results each time. Try it out for yourself.

```python
from sklearn.datasets.samples_generator import make_blobs

X,y = make_blobs(random_state=0)
```

Now we plot the data in a scatterplot. The color corresponds to the cluster, a point belongs to.

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

Let's do the same again, but this time we explicitly specify the cluster centers. Also we sample 200 points and set the standard deviation to 0.4. For detailed information on the parameters see the [documentation](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs)

```python
centers = [[1, 1], [-1, -1], [1, -1], [2,1]]
X,y = make_blobs(n_samples=200,centers=centers,cluster_std=0.4,random_state=0)

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

## Make_moons
This generator creates two interleaving half circles. This poses a problem to classifiers, that need the data to be linearly separable as for example logistic regression.
We create 200 sample points and add some noise to the data. Again we set the random_state to 0 for reproducible results. For details on the parameters see the [documentation](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html#sklearn.datasets.make_moons).

```python
from sklearn.datasets.samples_generator import make_moons

X,y = make_moons(n_samples=200,noise=0.1,random_state=0)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
